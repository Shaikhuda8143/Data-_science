# 2.Bivariate Analysis

```python
import pandas as pd
```

```python
mtcars_data=pd.read_csv('mtcars.csv')
mtcars_data
```

```python
mtcars_data.head()
```

###### To know 2 variables are directly related or indirectly related we use scatter plot

## 2.1 scatter plot- x & y must be continuous

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x='hp',y='mpg',data=mtcars_data)
plt.xlabel('hp',size=15)
plt.ylabel('mpg',size=15)
plt.title('MPG vs HP',size= 20)
plt.show()
```

```python
plt.scatter(x='wt',y='mpg',data=mtcars_data)
plt.xlabel('wt',size=15)
plt.ylabel('mpg',size=15)
plt.title('MPG vs WT',size= 20)
plt.show()
```

```python
plt.scatter(x='gear',y='mpg',data=mtcars_data)
plt.xlabel('gear',size=15)                                                   #not appropriate as we gave discreate data on x label
plt.ylabel('mpg',size=15)
plt.title('MPG vs gear',size= 20)
plt.show()
```

## 2.2 Bar plot

```python
mtcars_data.head()
```

### It takes x axis must be discreate & y axis must be continuous

```python
import seaborn as sns                                              #similar to count plot but in bar plot both x
                                                                      #& y must be given 
```

```python
sns.barplot(x='cyl', y='mpg',data=mtcars_data)
plt.xlabel('cyl',size=15)
plt.ylabel('mpg',size=15)
plt.title('CYL vs MPG',size=20)
plt.figure(figsize=(15,5))
plt.show()
```

# Multivariate analysis

```python
import seaborn as sns
```

```python
import pandas as pd
```

```python
mtcars_data=pd.read_csv('mtcars.csv')
mtcars_data
```

```python
mtcars_data.head()
```

```python
sns.pairplot(data=mtcars_data)
```

```python
mtcars_data_corr=mtcars_data.corr().round()
mtcars_data_corr
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,10))
sns.heatmap( data=mtcars_data_corr, annot=True)
```
